```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

# Set premium aesthetics
sns.set_theme(style="whitegrid")
plt.rcParams['figure.figsize'] = (10, 6)
```

```python
# Load the dataset
df = pd.read_csv('mobile1.csv')
print(" Dataset Preview:")
print(df.head())

# Independent variable (X)
X = df[['Screen On Time (hours/day)']]

# Dependent variable (y)
y = df['Battery Drain (mAh/day)']
```

```python
# Initialize and Train the Model
model = LinearRegression()
model.fit(X, y)

print(" Linear Regression model trained successfully.")
print(f"\nCoefficient (Screen On Time): {model.coef_[0]:.4f}")
print(f"Intercept: {model.intercept_:.4f}")
```

```python
# Visualize relationship
sns.regplot(x='Screen On Time (hours/day)', y='Battery Drain (mAh/day)', data=df, color='#e74c3c')
plt.title('Screen On Time vs Battery Drain')
plt.show()

# Custom Prediction with User Input
def predict_battery(screen_time):
    prediction = model.predict([[screen_time]])[0]
    print(f"\n Predicted Battery Drain:")
    print(f"Screen On Time: {screen_time}h")
    print(f"Estimated Battery Drain: {prediction:.2f} mAh/day")

try:
    u_screen = float(input("Enter screen on time (hours/day): "))
    predict_battery(u_screen)
except ValueError:
    print("❌ Please enter a numeric value.")
```

```python
# Model Evaluation
y_pred = model.predict(X)
accuracy = r2_score(y, y_pred)
print(f"\n Model R² Score: {accuracy:.4f}")
print("(Note: R² indicates how well Screen On Time explains Battery Drain variance)")
```
